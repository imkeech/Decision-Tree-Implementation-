```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/User/OneDrive/Documents/Bank.csv")
df.head(8)
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
dff=df.drop('Date',axis=1)
```

### Step 1

```python
df1=dff.sample(frac=0.5)
```

```python
df1
```

```python
X1=df1.drop('Adj Close',axis=1)
y1=df1.Close
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.2, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_trains = scaler.fit_transform(X_train)
X_tests = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

model = DecisionTreeRegressor(random_state=1)
model.fit(X_trains, y_train)
predicted_scaled = model.predict(X_tests)

r2 = r2_score(y_test, predicted_scaled)
print(f'R-squared with scaled features: {r2}')
```
